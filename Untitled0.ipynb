```python
# ============================================================
#  RaktCare AI — Complete ML Pipeline
#  Models:
#    1. Donor Availability Prediction (Classification)
#    2. Donation Frequency Prediction (Regression)
#    3. Blood Shortage Forecasting (Time-Series Regression)
#    4. Donor Ranking Engine
#    5. Blood Compatibility Lookup
# ============================================================

import pandas as pd
import numpy as np
import warnings
import json
from datetime import datetime, timedelta

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, RandomForestRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, classification_report, confusion_matrix,
    mean_absolute_error, mean_squared_error, r2_score
)
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.inspection import permutation_importance

import xgboost as xgb
import joblib

warnings.filterwarnings("ignore")
np.random.seed(42)


# ============================================================
# SECTION 1 — LOAD & EXPLORE DATA
# ============================================================

print("=" * 60)
print("RAKTCARE AI — ML PIPELINE")
print("=" * 60)

# Load datasets
registry_df   = pd.read_csv("blood_donation_registry_ml_ready.csv")
donor_df      = pd.read_csv("blood_donor_dataset.csv")
compat_df     = pd.read_csv("blood_compatibility_lookup.csv")

print(f"\n[1] Datasets Loaded")
print(f"    Registry (ML-ready)   : {registry_df.shape[0]:,} rows × {registry_df.shape[1]} cols")
print(f"    Donor Dataset         : {donor_df.shape[0]:,} rows × {donor_df.shape[1]} cols")
print(f"    Compatibility Lookup  : {compat_df.shape[0]} rows × {compat_df.shape[1]} cols")

print(f"\n[2] Target Distribution — donated_next_6m")
vc = registry_df["donated_next_6m"].value_counts()
print(f"    Will NOT donate (0) : {vc[0]:,}  ({vc[0]/len(registry_df)*100:.1f}%)")
print(f"    Will donate     (1) : {vc[1]:,}  ({vc[1]/len(registry_df)*100:.1f}%)")

print(f"\n[3] Missing Values")
missing = registry_df.isnull().sum()
print(missing[missing > 0].to_string() if missing.any() else "    No missing numeric values")


# ============================================================
# SECTION 2 — PREPROCESSING & FEATURE ENGINEERING
# ============================================================

print("\n" + "=" * 60)
print("SECTION 2 — PREPROCESSING & FEATURE ENGINEERING")
print("=" * 60)

df = registry_df.copy()

# --- Date features ---
df["last_donation_date"] = pd.to_datetime(df["last_donation_date"], errors="coerce")
df["as_of_date"]         = pd.to_datetime(df["as_of_date"], errors="coerce")

# Days since last donation (recency already present, but recompute cleanly)
df["days_since_last"]    = (df["as_of_date"] - df["last_donation_date"]).dt.days.fillna(df["recency_days"])

# Donation frequency: donations per year
df["donations_per_year"] = df["lifetime_donation_count"] / (df["years_since_first_donation"].replace(0, 1))

# Eligibility window: standard 56-day (3-month) gap in India
df["eligible_by_gap"]    = (df["days_since_last"] >= 90).astype(int)

# Donor loyalty score (0–1)
df["loyalty_score"]      = df["is_regular_donor"] * 0.5 + \
                           (df["donation_count_last_12m"] / (df["donation_count_last_12m"].max() + 1)) * 0.5

# Age buckets
df["age_bucket"] = pd.cut(
    df["age"],
    bins=[0, 25, 35, 45, 55, 65, 100],
    labels=["18-25", "26-35", "36-45", "46-55", "56-65", "65+"]
)

# BMI health flag
df["healthy_bmi"] = ((df["bmi"] >= 18.5) & (df["bmi"] <= 30)).astype(int)

# Encode categoricals
le_sex    = LabelEncoder()
le_region = LabelEncoder()
le_site   = LabelEncoder()
le_btype  = LabelEncoder()
le_elig   = LabelEncoder()
le_age_b  = LabelEncoder()

df["sex_enc"]    = le_sex.fit_transform(df["sex"].fillna("Unknown"))
df["region_enc"] = le_region.fit_transform(df["region"].fillna("Unknown"))
df["site_enc"]   = le_site.fit_transform(df["preferred_site"].fillna("Unknown"))
df["btype_enc"]  = le_btype.fit_transform(df["blood_type"].fillna("Unknown"))
df["elig_enc"]   = le_elig.fit_transform(df["eligibility_status"].fillna("Unknown"))
df["age_b_enc"]  = le_age_b.fit_transform(df["age_bucket"].astype(str))

FEATURES = [
    "age",
    "sex_enc",
    "is_rare_type",
    "smoker",
    "bmi",
    "chronic_condition_flag",
    "eligible_to_donate",
    "donation_count_last_12m",
    "is_regular_donor",
    "years_since_first_donation",
    "lifetime_donation_count",
    "recency_days",
    "days_since_last",
    "blood_type_country_prevalence",
    "donation_propensity_score",
    "region_enc",
    "site_enc",
    "btype_enc",
    "elig_enc",
    "donations_per_year",
    "eligible_by_gap",
    "loyalty_score",
    "healthy_bmi",
    "age_b_enc",
]

TARGET_CLF  = "donated_next_6m"
TARGET_REG  = "next_6m_donation_count"

X = df[FEATURES].fillna(0)
y_clf = df[TARGET_CLF]
y_reg = df[TARGET_REG]

print(f"\n    Feature matrix shape : {X.shape}")
print(f"    Features used        : {len(FEATURES)}")
print(f"\n    Engineered features added:")
print(f"      - days_since_last      (recomputed from dates)")
print(f"      - donations_per_year   (frequency metric)")
print(f"      - eligible_by_gap      (90-day gap rule)")
print(f"      - loyalty_score        (composite metric)")
print(f"      - healthy_bmi          (health flag)")


# ============================================================
# SECTION 3 — MODEL 1: DONOR AVAILABILITY PREDICTION
#              (Will this donor donate in the next 6 months?)
# ============================================================

print("\n" + "=" * 60)
print("SECTION 3 — MODEL 1: DONOR AVAILABILITY PREDICTION")
print("  Task: Binary Classification")
print("  Target: donated_next_6m (0 / 1)")
print("=" * 60)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_clf, test_size=0.2, random_state=42, stratify=y_clf
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# --- 3a. Logistic Regression (baseline) ---
print("\n  [3a] Logistic Regression (Baseline)")
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train_scaled, y_train)
lr_pred = lr.predict(X_test_scaled)
lr_proba = lr.predict_proba(X_test_scaled)[:, 1]

print(f"       Accuracy  : {accuracy_score(y_test, lr_pred):.4f}")
print(f"       Precision : {precision_score(y_test, lr_pred):.4f}")
print(f"       Recall    : {recall_score(y_test, lr_pred):.4f}")
print(f"       F1-Score  : {f1_score(y_test, lr_pred):.4f}")
print(f"       ROC-AUC   : {roc_auc_score(y_test, lr_proba):.4f}")

# --- 3b. Random Forest ---
print("\n  [3b] Random Forest Classifier")
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=12,
    min_samples_split=10,
    min_samples_leaf=4,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)
rf_pred  = rf.predict(X_test)
rf_proba = rf.predict_proba(X_test)[:, 1]

print(f"       Accuracy  : {accuracy_score(y_test, rf_pred):.4f}")
print(f"       Precision : {precision_score(y_test, rf_pred):.4f}")
print(f"       Recall    : {recall_score(y_test, rf_pred):.4f}")
print(f"       F1-Score  : {f1_score(y_test, rf_pred):.4f}")
print(f"       ROC-AUC   : {roc_auc_score(y_test, rf_proba):.4f}")

# --- 3c. XGBoost (primary model) ---
print("\n  [3c] XGBoost Classifier (Primary Model)")
scale_pos = (y_train == 0).sum() / (y_train == 1).sum()

xgb_clf = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos,
    eval_metric="logloss",
    use_label_encoder=False,
    random_state=42,
    n_jobs=-1
)
xgb_clf.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)
xgb_pred  = xgb_clf.predict(X_test)
xgb_proba = xgb_clf.predict_proba(X_test)[:, 1]

print(f"       Accuracy  : {accuracy_score(y_test, xgb_pred):.4f}")
print(f"       Precision : {precision_score(y_test, xgb_pred):.4f}")
print(f"       Recall    : {recall_score(y_test, xgb_pred):.4f}")
print(f"       F1-Score  : {f1_score(y_test, xgb_pred):.4f}")
print(f"       ROC-AUC   : {roc_auc_score(y_test, xgb_proba):.4f}")

print("\n  [3d] Detailed Classification Report (XGBoost)")
print(classification_report(y_test, xgb_pred, target_names=["Won't Donate", "Will Donate"]))

print("\n  [3e] Confusion Matrix (XGBoost)")
cm = confusion_matrix(y_test, xgb_pred)
print(f"       TN={cm[0,0]}  FP={cm[0,1]}")
print(f"       FN={cm[1,0]}  TP={cm[1,1]}")

# --- 3f. Cross-validation ---
print("\n  [3f] 5-Fold Stratified Cross-Validation (XGBoost)")
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(xgb_clf, X, y_clf, cv=cv, scoring="f1", n_jobs=-1)
print(f"       F1 per fold : {[f'{s:.4f}' for s in cv_scores]}")
print(f"       Mean F1     : {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")

# --- 3g. Feature Importance ---
print("\n  [3g] Top 10 Feature Importances (XGBoost)")
importance_df = pd.DataFrame({
    "Feature": FEATURES,
    "Importance": xgb_clf.feature_importances_
}).sort_values("Importance", ascending=False).head(10)

for _, row in importance_df.iterrows():
    bar = "█" * int(row["Importance"] * 300)
    print(f"       {row['Feature']:<35} {row['Importance']:.4f}  {bar}")


# ============================================================
# SECTION 4 — MODEL 2: DONATION FREQUENCY REGRESSION
#              (How many times will this donor donate in 6m?)
# ============================================================

print("\n" + "=" * 60)
print("SECTION 4 — MODEL 2: DONATION FREQUENCY REGRESSION")
print("  Task: Regression")
print("  Target: next_6m_donation_count")
print("=" * 60)

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

# --- Linear Regression baseline ---
print("\n  [4a] Linear Regression (Baseline)")
lin = LinearRegression()
lin.fit(X_train_r, y_train_r)
lin_pred = lin.predict(X_test_r)

print(f"       MAE  : {mean_absolute_error(y_test_r, lin_pred):.4f}")
print(f"       RMSE : {np.sqrt(mean_squared_error(y_test_r, lin_pred)):.4f}")
print(f"       R²   : {r2_score(y_test_r, lin_pred):.4f}")

# --- Random Forest Regressor ---
print("\n  [4b] Random Forest Regressor")
rf_reg = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)
rf_reg.fit(X_train_r, y_train_r)
rf_reg_pred = rf_reg.predict(X_test_r)

print(f"       MAE  : {mean_absolute_error(y_test_r, rf_reg_pred):.4f}")
print(f"       RMSE : {np.sqrt(mean_squared_error(y_test_r, rf_reg_pred)):.4f}")
print(f"       R²   : {r2_score(y_test_r, rf_reg_pred):.4f}")

# --- XGBoost Regressor (primary) ---
print("\n  [4c] XGBoost Regressor (Primary Model)")
xgb_reg = xgb.XGBRegressor(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)
xgb_reg.fit(X_train_r, y_train_r, eval_set=[(X_test_r, y_test_r)], verbose=False)
xgb_reg_pred = xgb_reg.predict(X_test_r)

print(f"       MAE  : {mean_absolute_error(y_test_r, xgb_reg_pred):.4f}")
print(f"       RMSE : {np.sqrt(mean_squared_error(y_test_r, xgb_reg_pred)):.4f}")
print(f"       R²   : {r2_score(y_test_r, xgb_reg_pred):.4f}")


# ============================================================
# SECTION 5 — MODEL 3: BLOOD SHORTAGE FORECASTING
#              (7-day demand forecast per blood group)
# ============================================================

print("\n" + "=" * 60)
print("SECTION 5 — MODEL 3: BLOOD SHORTAGE FORECASTING")
print("  Task: Simulate 7-day demand forecast per blood group")
print("=" * 60)

# Simulate daily demand data (in real deployment this comes from hospital DB)
np.random.seed(42)
blood_groups = ["A+", "A-", "B+", "B-", "O+", "O-", "AB+", "AB-"]
base_demand  = {"A+": 120, "A-": 30, "B+": 100, "B-": 25,
                "O+": 140, "O-": 20, "AB+": 40, "AB-": 10}

date_range = pd.date_range(end=datetime.today(), periods=90, freq="D")
demand_records = []
for date in date_range:
    for bg in blood_groups:
        # Simulate weekend dips, weekday spikes, some random noise
        base = base_demand[bg]
        wd   = date.weekday()
        seasonal_factor = 0.8 if wd >= 5 else 1.1
        noise = np.random.normal(0, base * 0.1)
        demand_records.append({
            "date":        date,
            "blood_group": bg,
            "units_needed": max(0, int(base * seasonal_factor + noise))
        })

demand_df = pd.DataFrame(demand_records)

# Build lag features for forecasting
def build_lag_features(df_bg):
    df_bg = df_bg.sort_values("date").copy()
    df_bg["lag_1"]  = df_bg["units_needed"].shift(1)
    df_bg["lag_3"]  = df_bg["units_needed"].shift(3)
    df_bg["lag_7"]  = df_bg["units_needed"].shift(7)
    df_bg["roll_7"] = df_bg["units_needed"].shift(1).rolling(7).mean()
    df_bg["weekday"] = df_bg["date"].dt.weekday
    df_bg["week"]    = df_bg["date"].dt.isocalendar().week.astype(int)
    return df_bg.dropna()

print("\n  Training one forecasting model per blood group:")
print(f"  {'Blood Group':<12} {'MAE':>8} {'RMSE':>8} {'R²':>8}  7-Day Forecast →")

forecast_results = {}
for bg in blood_groups:
    bg_df = demand_df[demand_df["blood_group"] == bg].copy()
    bg_df = build_lag_features(bg_df)

    feats = ["lag_1", "lag_3", "lag_7", "roll_7", "weekday", "week"]
    X_bg  = bg_df[feats]
    y_bg  = bg_df["units_needed"]

    split = int(len(X_bg) * 0.8)
    X_tr, X_te = X_bg.iloc[:split], X_bg.iloc[split:]
    y_tr, y_te = y_bg.iloc[:split], y_bg.iloc[split:]

    model_fg = xgb.XGBRegressor(n_estimators=100, max_depth=3,
                                 learning_rate=0.1, random_state=42)
    model_fg.fit(X_tr, y_tr, verbose=False)
    preds = model_fg.predict(X_te)

    mae  = mean_absolute_error(y_te, preds)
    rmse = np.sqrt(mean_squared_error(y_te, preds))
    r2   = r2_score(y_te, preds)

    # Simulate 7-day future forecast
    last_row = bg_df.iloc[-1]
    future_preds = []
    lag_vals = list(bg_df["units_needed"].values[-7:])
    for d in range(7):
        row = {
            "lag_1":   lag_vals[-1],
            "lag_3":   lag_vals[-3] if len(lag_vals) >= 3 else lag_vals[-1],
            "lag_7":   lag_vals[-7] if len(lag_vals) >= 7 else lag_vals[-1],
            "roll_7":  np.mean(lag_vals[-7:]),
            "weekday": (last_row["weekday"] + d + 1) % 7,
            "week":    last_row["week"]
        }
        pred = int(model_fg.predict(pd.DataFrame([row]))[0])
        future_preds.append(pred)
        lag_vals.append(pred)

    forecast_results[bg] = future_preds
    forecast_str = " | ".join([str(v) for v in future_preds])
    print(f"  {bg:<12} {mae:>8.2f} {rmse:>8.2f} {r2:>8.4f}  [{forecast_str}]")

print("\n  (Columns = Day 1 through Day 7 from today)")


# ============================================================
# SECTION 6 — BLOOD COMPATIBILITY ENGINE
# ============================================================

print("\n" + "=" * 60)
print("SECTION 6 — BLOOD COMPATIBILITY ENGINE")
print("=" * 60)

# Build lookup dict from CSV
compat_map = {}
for _, row in compat_df.iterrows():
    key = (row["donor_blood_type"], row["recipient_blood_type"])
    compat_map[key] = {
        "compatible": bool(row["compatible_for_rbc_transfusion"]),
        "level":      row["compatibility_level"]
    }

def check_compatibility(donor_type: str, recipient_type: str) -> dict:
    """Returns compatibility info between a donor and recipient blood type."""
    result = compat_map.get((donor_type, recipient_type))
    if result is None:
        return {"compatible": False, "level": "unknown", "note": "Type not in lookup"}
    return {
        "donor":      donor_type,
        "recipient":  recipient_type,
        "compatible": result["compatible"],
        "level":      result["level"]
    }

def get_compatible_donors(recipient_type: str) -> list:
    """Returns all donor types compatible with a given recipient."""
    compatible = []
    for (donor, recip), info in compat_map.items():
        if recip == recipient_type and info["compatible"]:
            compatible.append({"donor_type": donor, "level": info["level"]})
    compatible.sort(key=lambda x: (x["level"] != "ideal", x["level"]))
    return compatible

# Demo
print("\n  Sample Compatibility Checks:")
tests = [("O-", "AB+"), ("AB+", "O-"), ("A+", "A+"), ("B-", "AB-")]
for d, r in tests:
    c = check_compatibility(d, r)
    symbol = "✅" if c["compatible"] else "❌"
    print(f"    {symbol}  Donor {d:4s} → Recipient {r:4s}  [{c['level']}]")

print("\n  Compatible donors for recipient type O-:")
for entry in get_compatible_donors("O-"):
    print(f"    • {entry['donor_type']:4s}  [{entry['level']}]")


# ============================================================
# SECTION 7 — DONOR RANKING ENGINE
#              (Core of the SOS Broadcast feature)
# ============================================================

print("\n" + "=" * 60)
print("SECTION 7 — DONOR RANKING ENGINE")
print("  Simulates real-time ranking for SOS Broadcast")
print("=" * 60)

def rank_donors_for_sos(
    recipient_blood_type: str,
    available_donors_df: pd.DataFrame,
    model,
    feature_cols: list,
    top_n: int = 10
) -> pd.DataFrame:
    """
    Given a recipient blood type and a pool of donors, returns the
    top-N donors ranked by:
      1. Blood type compatibility
      2. ML-predicted availability probability
      3. Eligibility (donation gap, health flags)
      4. Recency (prefer donors who haven't donated recently)
    """
    # Step 1: Filter by compatibility
    compatible_types = [
        e["donor_type"] for e in get_compatible_donors(recipient_blood_type)
    ]
    pool = available_donors_df[
        available_donors_df["blood_type"].isin(compatible_types)
    ].copy()

    if pool.empty:
        print(f"    No compatible donors found for {recipient_blood_type}")
        return pd.DataFrame()

    # Step 2: Encode & predict availability probability
    pool_X = pool[feature_cols].fillna(0)
    pool["availability_prob"] = model.predict_proba(pool_X)[:, 1]

    # Step 3: Compatibility score (ideal=1.0, acceptable=0.7)
    compat_level = {
        (donor, recipient_blood_type): info["level"]
        for (donor, recipient), info in compat_map.items()
        if recipient == recipient_blood_type and info["compatible"]
    }
    pool["compat_score"] = pool["blood_type"].apply(
        lambda bt: 1.0 if compat_level.get((bt, recipient_blood_type)) == "ideal" else 0.7
    )

    # Step 4: Eligibility score
    pool["eligibility_score"] = pool["eligible_to_donate"].astype(float)

    # Step 5: Recency score (longer since last donation = better, capped at 365 days)
    pool["recency_score"] = (pool["recency_days"].clip(0, 365) / 365).fillna(0)

    # Step 6: Composite RaktScore
    pool["rakt_score"] = (
        0.40 * pool["availability_prob"] +
        0.25 * pool["compat_score"] +
        0.20 * pool["eligibility_score"] +
        0.15 * pool["recency_score"]
    )

    # Step 7: Sort and return top N
    ranked = pool.sort_values("rakt_score", ascending=False).head(top_n).reset_index(drop=True)
    ranked["rank"] = ranked.index + 1

    return ranked[[
        "rank", "donor_id", "blood_type", "age", "region",
        "availability_prob", "compat_score", "eligibility_score",
        "recency_score", "rakt_score"
    ]]

# Demo: simulate SOS for O- patient
print("\n  Simulating SOS Broadcast for recipient blood type: O-")
sample_pool = df.sample(500, random_state=7).copy()
ranked_donors = rank_donors_for_sos("O-", sample_pool, xgb_clf, FEATURES, top_n=10)

print(f"\n  Top 10 donors ranked (out of {len(sample_pool)} in pool):\n")
print(f"  {'#':<4} {'ID':<12} {'Type':<6} {'Age':<5} {'Region':<10} "
      f"{'Avail%':<8} {'Compat':<8} {'Eligible':<10} {'RaktScore'}")
print("  " + "-" * 80)
for _, row in ranked_donors.iterrows():
    print(
        f"  {int(row['rank']):<4} {str(row['donor_id']):<12} {row['blood_type']:<6} "
        f"{int(row['age']):<5} {str(row['region']):<10} "
        f"{row['availability_prob']*100:>6.1f}%  "
        f"{row['compat_score']:>6.2f}   "
        f"{'Yes' if row['eligibility_score'] else 'No':<10} "
        f"{row['rakt_score']:.4f}"
    )


# ============================================================
# SECTION 8 — DONOR HEALTH PASSPORT DATA BUILDER
# ============================================================

print("\n" + "=" * 60)
print("SECTION 8 — DONOR HEALTH PASSPORT DATA BUILDER")
print("=" * 60)

def generate_passport(donor_row: pd.Series) -> dict:
    """Generate the data payload for a donor's digital health passport."""
    eligibility_status = "ELIGIBLE" if donor_row["eligible_to_donate"] == 1 else "DEFERRED"
    last_date = donor_row.get("last_donation_date", "Unknown")
    if pd.notnull(last_date):
        last_date_str = str(last_date)[:10]
        next_eligible  = (pd.to_datetime(last_date) + timedelta(days=90)).strftime("%b %d, %Y")
    else:
        last_date_str = "N/A"
        next_eligible  = "N/A"

    return {
        "donor_id":           donor_row["donor_id"],
        "blood_type":         donor_row["blood_type"],
        "age":                donor_row["age"],
        "total_donations":    donor_row["lifetime_donation_count"],
        "last_donated":       last_date_str,
        "next_eligible":      next_eligible,
        "status":             eligibility_status,
        "is_rare_type":       bool(donor_row["is_rare_type"]),
        "rakt_score":         round(xgb_clf.predict_proba(
                                  pd.DataFrame([donor_row[FEATURES].fillna(0)])
                              )[0, 1], 4),
        "generated_at":       datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }

sample_passport = generate_passport(df.iloc[0])
print("\n  Sample Passport (Donor #1):")
for k, v in sample_passport.items():
    print(f"    {k:<22} : {v}")


# ============================================================
# SECTION 9 — MODEL EVALUATION SUMMARY
# ============================================================

print("\n" + "=" * 60)
print("SECTION 9 — FINAL MODEL EVALUATION SUMMARY")
print("=" * 60)

metrics = {
    "Model 1 — Availability (XGBoost CLF)": {
        "Accuracy":  round(accuracy_score(y_test, xgb_pred), 4),
        "Precision": round(precision_score(y_test, xgb_pred), 4),
        "Recall":    round(recall_score(y_test, xgb_pred), 4),
        "F1-Score":  round(f1_score(y_test, xgb_pred), 4),
        "ROC-AUC":   round(roc_auc_score(y_test, xgb_proba), 4),
    },
    "Model 2 — Frequency (XGBoost REG)": {
        "MAE":  round(mean_absolute_error(y_test_r, xgb_reg_pred), 4),
        "RMSE": round(np.sqrt(mean_squared_error(y_test_r, xgb_reg_pred)), 4),
        "R²":   round(r2_score(y_test_r, xgb_reg_pred), 4),
    },
}

for model_name, scores in metrics.items():
    print(f"\n  {model_name}")
    for metric, val in scores.items():
        print(f"    {metric:<12} : {val}")

print("\n  Shortage Forecasting: Per-blood-group XGBoost regressors")
print("  Compatibility Engine: Rule-based lookup (64-entry table)")
print("  Donor Ranking:        Composite RaktScore (4-factor formula)")


# ============================================================
# SECTION 10 — SAVE MODELS
# ============================================================

print("\n" + "=" * 60)
print("SECTION 10 — SAVING MODELS & ARTIFACTS")
print("=" * 60)

joblib.dump(xgb_clf,    "raktcare_availability_model.pkl")
joblib.dump(xgb_reg,    "raktcare_frequency_model.pkl")
joblib.dump(scaler,     "raktcare_scaler.pkl")
joblib.dump(compat_map, "raktcare_compat_map.pkl")

model_meta = {
    "project":          "RaktCare AI",
    "version":          "1.0.0",
    "trained_on":       str(datetime.now().date()),
    "train_samples":    len(X_train),
    "test_samples":     len(X_test),
    "features":         FEATURES,
    "availability_f1":  round(f1_score(y_test, xgb_pred), 4),
    "availability_auc": round(roc_auc_score(y_test, xgb_proba), 4),
    "frequency_r2":     round(r2_score(y_test_r, xgb_reg_pred), 4),
}
with open("raktcare_model_meta.json", "w") as f:
    json.dump(model_meta, f, indent=2)

print("\n  Saved:")
print("    raktcare_availability_model.pkl")
print("    raktcare_frequency_model.pkl")
print("    raktcare_scaler.pkl")
print("    raktcare_compat_map.pkl")
print("    raktcare_model_meta.json")

print("\n" + "=" * 60)
print("  ✅  RaktCare AI ML Pipeline Complete")
print("      Every Drop. Every Life.")
print("=" * 60)
```
